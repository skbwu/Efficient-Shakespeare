```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import os, pickle
from matplotlib.lines import Line2D
from matplotlib.patches import Patch

# colorblind friendly colors
colors = ['#377eb8', '#ff7f00', '#4daf4a',
          '#f781bf', '#a65628', '#984ea3',
          '#999999', '#e41a1c', '#dede00']
```

# Preprocessing

```python
# 1. GPT2-XL scores for word-based models
with open("results/scores.pickle", "rb") as file:
    scores = pickle.load(file)
    
# 2. GPT2-XL scores for BASELINE character-based models
with open("results/baseline_scores.pickle", "rb") as file:
    baseline_scores = pickle.load(file)
    
# 3. model information of word-based models
logs = pd.read_csv("results/logs.csv").fillna("None")
logs.freeze_type = logs.freeze_type.astype(str)
    
# 4. model information of BASELINE models
baseline_logs = pd.read_csv("results/baseline_logs.csv")
```

```python
# combine our scores + baseline scores into one dictionary
all_scores = {**baseline_scores, **scores}

# get question-wise mean and standard-deviations for standardization
prompt_means = np.array([all_scores[key] for key in all_scores.keys()]).mean(axis=0)
prompt_sds = np.array([all_scores[key] for key in all_scores.keys()]).std(axis=0)

# standardize all scores
for key in all_scores.keys():
    all_scores[key] = (np.array(all_scores[key]) - prompt_means) / prompt_sds
```

# Results Figure 1: Embedding Size vs. Performance Across Architectures

```python
# 3x1 grid of subplots
fig, ax = plt.subplots(1, 3, dpi=300, figsize=(13, 3.5), sharey=True)

# fixed values as our x-axis for all subplots
embed_sizes = [192, 384, 576, 768, 960]

# freeze_type = {True, False, None} vs. baselines
for i, freeze_type in enumerate([True, False, None]):
    
    # put in the curves for each vocabulary-size / mc
    for j, mc in enumerate([1, 3, 5]):
        
        # store our values for each embed_size (25%, 50%, 75% quantiles)
        low, med, high = [], [], []
        for embed_size in embed_sizes:
            low.append(np.quantile(a=all_scores[(mc, embed_size, freeze_type)], q=0.40))
            med.append(all_scores[(mc, embed_size, freeze_type)].mean())
            high.append(np.quantile(a=all_scores[(mc, embed_size, freeze_type)], q=0.60))
        
        # make our plots - point estimates with MEAN + uncertainty quantification
        ax[i].plot(embed_sizes, med, color=colors[j], marker="o")
        ax[i].fill_between(embed_sizes, low, high, alpha=0.3, color=colors[j])
    
    # add in our baseline, too
    baseline_low = [np.quantile(all_scores[embed_size], q=0.40) for embed_size in embed_sizes]
    baseline_high = [np.quantile(all_scores[embed_size], q=0.60) for embed_size in embed_sizes]
    ax[i].plot(embed_sizes, [all_scores[embed_size].mean() for embed_size in embed_sizes], 
               marker="^", color="grey")
    ax[i].fill_between(embed_sizes, baseline_low, baseline_high, color="grey", alpha=0.3)
    
    # beautify
    ax[i].set_title(["Frozen Word2Vec", "Warm-Start Word2Vec", "Random"][i], 
                    fontweight="bold", fontsize=14)
    ax[i].grid()
    ax[i].set_xlabel("Embedding Size", fontsize=13)
    if i == 0:
        ax[i].set_ylabel("GPT2-XL Eval. Score", fontsize=13)
    
# final beautifying
custom_lines = [
    Line2D([0], [0], color=colors[0], marker="o", label="Vocab Size: 13332 (Full)"),
    Line2D([0], [0], color=colors[1], marker="o", label="Vocab Size: 5338"),
    Line2D([0], [0], color=colors[2], marker="o", label="Vocab Size: 3600"),
    Line2D([0], [0], color="grey", marker="^", label="Baseline Char. Model")
]
fig.legend(handles=custom_lines, handlelength=1.0,
           loc="lower center", ncol=4, fontsize=13, bbox_to_anchor=(0.525, -0.1))
plt.tight_layout()
plt.show()
```

# Results Figure 2: Computational Costs vs. Performance Scatterplot

```python
# grid of subplots
fig, ax = plt.subplots(dpi=300, figsize=(13, 3.75), sharey=True)

# colors for vocabulary size, markers for freeze-type
markers = ["*", "x", "d"]

# plot 1: trainable parameter count vs. median log-prob
# plot 2: train time (per epoch, seeconds) vs. median log-prob
for i, mc in enumerate([1, 3, 5]):
    for embed_size in embed_sizes:
        for j, freeze_type in enumerate([True, False, None]):
            
            # get the median log-prob (standardized) + the parameter count (in millions)
            med_log_prob = np.median(all_scores[(mc, embed_size, freeze_type)])
            param_count_million = logs.query(
                f"minimum_count == {mc} and embed_size == {embed_size} " + 
                f"and freeze_type == '{str(freeze_type)}'").trainable_params / 1e6
            
            # scatter these metrics on their relevant plots + annotate
            ax.scatter(param_count_million, med_log_prob, 
                          color=colors[i], marker=markers[j], s=70)
            ax.annotate(str(embed_size), (param_count_million.iloc[0], med_log_prob), fontsize=11)
            
# scatter the baseline values, too
ax.plot(baseline_logs.trainable_params.values / 1e6, 
              [np.median(all_scores[embed_size]) for embed_size in embed_sizes], 
              color="grey", marker="^", linestyle="--")

# annotate the baseline models, too
for embed_size in embed_sizes:
    ax.annotate(
        str(embed_size), 
        (baseline_logs.query(f"embed_size == {embed_size}").trainable_params.iloc[0] / 1e6, 
         np.median(all_scores[embed_size])), fontsize=11)

# labeling + beautifying
ax.set_xlabel("Parameter Count (Millions)", fontsize=14)
ax.set_ylabel("GPT2-XL Eval. Score", fontsize=14)
ax.set_title("Parameter Count vs. Performance", fontweight="bold", fontsize=16)
ax.tick_params("both", labelsize=12)
ax.grid()
custom_lines = [
    Line2D([0], [0], color=colors[0], marker="o", label="Vocab Size: 13332 (Full)"),
    Line2D([0], [0], color="grey", marker="*", label="Frozen Word2Vec"),
    Line2D([0], [0], color=colors[1], marker="o", label="Vocab Size: 5338"),
    Line2D([0], [0], color="grey", marker="x", label="Warm-Start Word2Vec"),
    Line2D([0], [0], color=colors[2], marker="o", label="Vocab Size: 3600"),
    Line2D([0], [0], color="grey", marker="d", label="Random"),
    Line2D([0], [0], color="grey", marker="^", linestyle="--", label="Baseline Char. Model")
]
fig.legend(handles=custom_lines, handlelength=1.0,
           loc="lower center", ncol=4, fontsize=12, bbox_to_anchor=(0.525, -0.15))
plt.tight_layout()
plt.show()
```

# Discussion / Appendix Figure - Regarding Warm-Start vs. Frozen W2V

```python
# dictionary to store our 95% quantiles of percent change in learned embeddings
percent_diffs = {}

# go thru all of our settings
for embed_size in [192, 384, 576, 768, 960]:
    for mc in [1, 3, 5]:
        
        # load in our models + set to eval mode
        model_true = torch.load(f"models/mc={mc}_embed-size={embed_size}_freeze-type=True/model.pth")
        model_false = torch.load(f"models/mc={mc}_embed-size={embed_size}_freeze-type=False/model.pth")
        model_true.eval(); model_false.eval()
        
        # compute the % change in each embedding vector
        with torch.no_grad():
            
            # get the difference of the norms, and the norm of the original
            norm_diff = torch.linalg.norm(
                model_true.token_embedding_table.weight\
                - model_false.token_embedding_table.weight, axis=1)
            norm_og = torch.linalg.norm(model_true.token_embedding_table.weight, axis=1)
            
            # compute the difference + report the 95% quantile + add to our list
            med_percent_diff = np.quantile((norm_diff / norm_og).cpu().numpy(), q=0.95)
            percent_diffs[(mc, embed_size)] = med_percent_diff
```

```python
# create our barplot figure
fig, ax = plt.subplots(dpi=300, figsize=(11, 3))
for i, embed_size in enumerate([192, 384, 576, 768, 960]):
    
    # do our bar plots
    plt.bar(x=[i-0.2], height=[percent_diffs[1, embed_size]], width=0.2, color=colors[0])
    plt.bar(x=[i], height=[percent_diffs[3, embed_size]], width=0.2, color=colors[1])
    plt.bar(x=[i+0.2], height=[percent_diffs[5, embed_size]], width=0.2, color=colors[2])

# beautify at the end
ax.set_xticks([0, 1, 2, 3, 4])
ax.set_xticklabels(labels=["$D=192$", "$D=384$", "$D=576$", "$D=768$", "$D=960$"])
plt.xlabel("Embedding Dimension Size", fontsize=13)
plt.ylabel("% Change Per Vector", fontsize=13)
plt.title("95th Percentile Changes in Embedding Vectors: Warm-Start vs. Frozen W2V", 
          fontweight="bold", fontsize=14)
plt.grid()
custom_lines = [
    Patch(color=colors[0], label="Vocab Size: 13332 (Full)"),
    Patch(color=colors[1], label="Vocab Size: 5338"),
    Patch(color=colors[2], label="Vocab Size: 3600"),
]
fig.legend(handles=custom_lines, handlelength=1.0,
           loc="lower center", ncol=4, fontsize=12, bbox_to_anchor=(0.525, -0.15))
plt.tight_layout()
plt.show()
```
