```python
from gensim.models import Word2Vec
import numpy as np
import re, os
import matplotlib.pyplot as plt
from collections import Counter
import scipy.stats as stats
import pandas as pd
import time
from IPython.display import clear_output
import seaborn as sns
from tqdm.notebook import tqdm
```

# Tokenizing Shakespeare Corpus into Words + Special Characters.

```python
with open("input.txt", "rt") as file:
    raw_corpus = file.readlines()
    
# special characters that we do want to keep!
spc = r'\w+|[^\w\s]|[\n\t\r\f\v]'

# get our corpus in splitted form for word2vec
corpus = [re.findall(spc, line) for line in raw_corpus]

# combine into one big list for aggregate statistics
combined = []
for line in corpus:
    combined += line
    
# how many times does each word appear in our corpus?
counts_dict = dict(Counter(combined))
counts = np.array(list(counts_dict.values()))

# separate alpha-numeric vs. special characters
special_tokens = [token for token in counts_dict.keys() if not token.isalpha()]
alpha_tokens = [token for token in counts_dict.keys() if token.isalpha()]
alpha_token_counts = np.array([counts_dict[token] for token in alpha_tokens])

# compute average number of alpha-numeric tokens per sentence
sentence_lengths = np.array([len([word for word in sentence if word.isalpha()]) for sentence in corpus])
sentence_lengths = sentence_lengths[sentence_lengths != 0]
```

# Exploratory Data Analysis towards Determining Word2Vec Parameters.

```python
# creating Figure 1 + Figure 2
fig, ax = plt.subplots(1, 2, dpi=300, figsize=(10, 3.5))

# Figure 1: percent of corpus captured by our alphabetical tokens + OVERLAY: counts of each token.
ax[0].plot(np.sort(alpha_token_counts)[::-1].cumsum() / alpha_token_counts.sum())
ax[0].axhline(y=0.90, linestyle="--", color="red"); ax[0].axvline(x=3000, linestyle="--", color="grey")
ax[0].tick_params("both", labelsize=9)
ax[0].set_title("Prop. of Corpus Captured by Tokens", fontweight="bold", fontsize=12)
ax[0].set_xlabel("Number of Tokens Used", fontsize=12)
ax[0].set_ylabel("Prop. of Corpus Captured", fontsize=12)
ax2 = ax[0].twinx()
ax2.fill_between(
    np.arange(len(alpha_token_counts)), 
    np.sort(alpha_token_counts)[::-1], color="grey", alpha=0.3)
ax2.set_ylim(bottom=0, top=20)
ax2.set_ylabel("# of Appearances of Token", fontsize=12)
ax2.grid(linestyle="-.")

# Figure 2: histogram of average sentence lengths in our corpus
ax[1].hist(sentence_lengths, bins=15, alpha=0.5, density=True)
ax[1].set_xlabel("Number of Tokens in Sentence", fontsize=12)
ax[1].grid(); ax[1].set_ylabel("Frequency of Sentences", fontsize=12)
ax[1].set_title("Distribution of Sentence Lengths", fontweight="bold", fontsize=12)
for q in [0.5, 0.75, 0.9, 0.95, 0.99]:
    ax[1].axvline(x=np.quantile(a=sentence_lengths, q=q), linestyle="--", color="red")
    
# beautify at the end
plt.tight_layout()
plt.show()
```

# Training Candidate Word2Vec Models.

```python
# create a directory to store our models (can embed into nn.Embedding later and then freeze)
if "word2vec_models" not in os.listdir():
    os.mkdir("word2vec_models")

# fix the number of epochs
epochs = 30

# training many word2vec models - directly prop. to prop of corpus captured
for min_count in [1, 2, 3, 4, 5, 6]:
    for vector_size in [192, 384, 576, 768, 960, 1152]:
        
        # train the model with these settings
        model = Word2Vec(
            sentences=corpus, 
            vector_size=vector_size, 
            window=12, min_count=min_count, workers=4, epochs=30)
        
        # save our model
        model.save(f"word2vec_models/mc={min_count}_vs={vector_size}.model")
        
        # status update
        clear_output(wait=True)
        print(f"Finished training model with min_count={min_count}, vector_size={vector_size}.")
```

# Visualizing Cosine Similarity Matrices of Entire Vocabularies

```python
# let's compare cosine similarities of the top tokens
fig, ax = plt.subplots(5, 3, dpi=300, figsize=(6, 10))

# get the words that are common to all of our embeddings
baseline_words = list(Word2Vec.load(f"word2vec_models/mc=6_vs=192.model").wv.key_to_index.keys())

for j in tqdm(range(3)):
    min_count = [1, 3, 5][j]
    
    for i in tqdm(range(5)):
        
        vector_size = [192, 384, 576, 768, 960][i]
        
        # load in the model
        model = Word2Vec.load(f"word2vec_models/mc={min_count}_vs={vector_size}.model")
        
        # compute the cosine similarities of these embeddings
        X = np.array([model.wv[word] for word in model.wv.index_to_key])
        cossims = (X @ X.T) / \
        (np.linalg.norm(X, axis=1).reshape(-1, 1)\
         * np.linalg.norm(X, axis=1).reshape(1, -1))
        
        # plot the cosine similarities
        sns.heatmap(cossims, cmap="bwr", ax=ax[i, j], cbar=False)
        ax[i, j].set_xticks([]); ax[i, j].set_yticks([])
        
        if i == 0:
            num_tokens = len(model.wv.key_to_index)
            ax[i, j].set_title(f"MC: {min_count}, # Tokens: {num_tokens}", fontsize=8)
        if j == 0:
            ax[i, j].set_ylabel(f"Emb. Size: {vector_size}", fontsize=8)   

# beautify
plt.tight_layout()
plt.savefig("cossim_matrices.png", facecolor="white", bbox_inches="tight")
plt.close()
```

# Cosine Similarities of Select Words on Largest Vocabulary (Sanity-Check)

```python
# load in the model
model = Word2Vec.load(f"word2vec_models/mc=1_vs=960.model")

# what words are we interested in?
words = [":", "the", "a", "Romeo", "Juliet", "God", "pray", "love", "war", "quickly"]

# create our cosine similarity matrix
cossim = np.zeros((len(words), len(words)))
for i, word1 in enumerate(words):
    for j, word2 in enumerate(words):
        
        v1, v2 = model.wv[word1], model.wv[word2]
        cossim[i, j] = (v1 @ v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
        
# create our heatmap + show the image
sns.heatmap(cossim, annot=True, xticklabels=words, yticklabels=words)
plt.title("Sample Cosine Similarities of Terms", fontweight="bold")
plt.tight_layout(); plt.show()
```
